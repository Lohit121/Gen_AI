## List Methods 
#### 1.append() --> insert the value at the end of the list 
#### 2.insert() --> insert the value at the specified index
#### 3.remove() --> remove the first occurrence of the element from the list
#### 4.pop() --> remove the element at the specified index and return it
#### 5.count() --> return the number of occurrences of the element in the list
#### 6.index() --> return the index of the first occurrence of the element in the list   
#### 7.sort() --> sort the list in ascending order
#### 8.sort(reverse=True) --> sort the list in descending order
#### 9.reverse() --> reverse the order of the list
#### 10.clear() --> remove all the elements from the list

## List comprehension 

### We are going to use looping statements inside the list so that we can store the values inside the list ina an iterable format and in list coprehension conditional statements are allowed as well

## Syntax for List comprehension is 

### [expression for item in range if condition]

### expression ---> what we want to print 
### for item in range ---> looping statement 
### if condition ---> conditional statement 
 

```python
# I need to store around values from 1 to 50 inside the list 
```

```python
list_items=[x for x in range(1,51)]
list_items
```

```python
list_square = [x**2 for x in range(1,11)]
```

```python
list_square
```

```python
#wanted to store only even square values 
```

```python
list_even_square = [x**2 for x in range(1,11) if x%2==0]
```

```python
list_even_square
```
