# Tuples

### Tuples is data structure that can store data in ordered way just like list but tuples are not mutable  

### Tuples will be created inside the round brackets ()

```python
a=tuple()
```

```python
type(a)
```

```python
tuple_values = (1,2,3,4,5)
tuple_values
```

#### Tuples are just like lists which can store data inside it in ordered way(indexed) but are going to be immutable

```python
tuple_values[1]
```

```python
tuple_values[1]=23
```

### Tuples are also capable of storing data of any data type 

```python
tuples_data = (1,2,"lohit",3.4)
```

```python
tuples_data
```

```python
tuples_data[2]
```

```python
tuples_data[:]
```

```python
tuples_data[1:]
```

```python
tuples_data[::-1]
```

```python
tuples_data[1:10]
```

### Methods in Tuples 

## Tuples , we are going to have 2 methods usually as :
## 1.count()
## 2.index()

```python
tuple_methods=(1,3,1,5,6,8,1,1,3,5,6,7,8,9)
```

```python
tuple_methods.index(3)
```

```python
tuple_methods.count(1)
```

### Packing and unpacking of Tuples

```python
# packing of tuples -> if you declare a variable with multiple values, it will be packed as a tuple by default

a=1,2,3,4,5,6,7,8,9
```

```python
print(a)
```

```python
type(a)
```

```python
# unpacking of tuples -> if you have a tuple with multiple values, you can unpack it into multiple variables
a,b,c,d=1,2,3,4
```

```python
a
```

```python
d
```

```python
type(d)
```

```python
## Unpacking with a *(star)

first_value,*middle_value,final_value=(1,2,3,4,5,6,7,8,9)
```

```python
middle_value
```

```python
type(middle_value)
```

```python
first_value,final_value
```

```python
type(first_value)
```
